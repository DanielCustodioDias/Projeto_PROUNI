```python
# Criar um DataFrame auxiliar para contar as ocorrências de cada parâmetro por ano
count_df = df.groupby(['ano', 'porcentagem_bolsa']).size().unstack(fill_value=0)

# Plotar o gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(10, 5))

# Definir a largura das barras agrupadas
bar_width = 0.25

# Definir os índices para a posição das barras
index = count_df.index

# Plotar as barras para cada parâmetro
ax.bar(index - bar_width, count_df['BOLSA INTEGRAL'], width=bar_width, label='BOLSA INTEGRAL', color='DarkCyan')
ax.bar(index, count_df['BOLSA PARCIAL 50%'], width=bar_width, label='BOLSA PARCIAL 50%', color='DarkSlateGray')

# Configurar os rótulos do eixo X, Y e o título
ax.set_xlabel('Ano')
ax.set_ylabel('Porcentagem Bolsa')
ax.set_title('Comparação ano com porcentagem de bolsa')
ax.set_xticks(index)
ax.set_xticklabels(index)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Filtrar as linhas onde 'porcentagem_bolsa' é igual a 'BOLSA INTEGRAL'
filtered_df = df[df['porcentagem_bolsa'] == 'BOLSA INTEGRAL']

# Agrupar o DataFrame filtrado por ano e calcular a soma da quantidade
grouped_df = filtered_df.groupby('ano')['porcentagem_bolsa'].count()

# Plotar o gráfico de linha
fig, ax = plt.subplots(figsize=(10, 5))

# Plotar a linha
ax.plot(grouped_df.index, grouped_df.values, marker='o', color='DarkCyan')

# Configurar os rótulos do eixo X, Y e o título
ax.set_xlabel('Ano')
ax.set_ylabel('Quantidade de Bolsas Integrais')
ax.set_title('Soma da Quantidade de Bolsas Integrais por Ano')

plt.tight_layout()
plt.show()
```

```python
# Filtrar as modalidades de bolsa desejadas
modalidades = ['BOLSA INTEGRAL', 'BOLSA PARCIAL 50%']
filtered_df = df[df['porcentagem_bolsa'].isin(modalidades)]

# Agrupar o DataFrame filtrado por 'instituto' e 'porcentagem_bolsa' e calcular a contagem
grouped_df = filtered_df.groupby(['instituto', 'porcentagem_bolsa']).size().unstack(fill_value=0)

# Calcular a soma total de bolsistas para cada instituto
grouped_df['Total'] = grouped_df.sum(axis=1)

# Ordenar o DataFrame pelo total de bolsistas em ordem decrescente
sorted_df = grouped_df.sort_values(by='Total', ascending=False).head(10)
# Plotar o gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(10, 6))

# Plotar as barras
sorted_df[modalidades].plot(kind='bar', stacked=True, ax=ax)

# Configurar os rótulos do eixo X, Y e o título
ax.set_xlabel('Instituto')
ax.set_ylabel('Quantidade de Bolsas')
ax.set_title('Quantidade de Bolsas por Modalidade e Instituto')
ax.legend(title='Modalidade')

plt.tight_layout()
plt.show()
```
